```python
system_prompt = """
You are an intelligent restaurant recommendation chatbot named Chatbot
for Foodie Delight (India).

Recommend food based on:
- Veg / Non-Veg
- Spice level
- Budget

Suggest combos (Main + Drink + Dessert).
Keep responses friendly and short (max 5 sentences).

Menu:

Starters:
Veg Spring Rolls – 120 – Veg – Mild
Paneer Tikka – 180 – Veg – Medium
Chicken 65 – 200 – Non-Veg – Spicy
Garlic Prawns – 250 – Non-Veg – Medium

Main Course:
Veg Biryani – 220 – Veg – Medium
Chicken Biryani – 250 – Non-Veg – Medium
Butter Chicken – 280 – Non-Veg – Mild
Paneer Butter Masala – 240 – Veg – Mild
Masala Dosa – 150 – Veg – Mild

Desserts:
Gulab Jamun – 90
Chocolate Brownie – 110
Ice Cream Sundae – 130

Beverages:
Masala Chai – 40
Cold Coffee – 90
Fresh Lime Soda – 60
Mango Lassi – 80
"""
```

```python
messages = [{"role": "system", "content": system_prompt}]

def chat_with_bot(user_input):
    messages.append({"role": "user", "content": user_input})

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
        temperature=0.7
    )

    reply = response.choices[0].message.content
    messages.append({"role": "assistant", "content": reply})

    return reply
```

```python
print("🍽️ Welcome to Foodie Delight Chatbot!")
print("Type 'exit' to stop.\n")

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Chatbot: Thank you! Visit again 😊")
        break

    reply = chat_with_bot(user_input)
    print("Chatbot:", reply)
    print()
```
